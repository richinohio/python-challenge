```python
import gmaps
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import api_keys
import json
from citipy import citipy
```

```python
lat_lngs= []
```

```python
lats= np.random.uniform(low=-90, high= 90, size=1500)
```

```python
lngs= np.random.uniform(low=-180, high= 180, size=1500)
```

```python
lat_lngs= zip(lats, lngs)
```

```python
cities =[]
```

```python
for lat_lng in lat_lngs:
    city= citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
api_key= "1cd626ed596fd71f1b514955bb362e2e"
```

```python
url= "http://api.openweathermap.org/data/2.5/weather?units-Imperial&APPID=" + api_key
```

```python
city_data= []
```

```python
for city in cities:
    query_url= url + "&q=" + city
    print (f"Working on {city}")
    try:
        response = requests.get(query_url).json()
        city_lat= response["coord"]["lat"]
        city_lng= response["coord"]["lon"]
        city_max_temp= response["main"]["temp_max"]
        city_humidity= response["main"]["humidity"]
        city_wind= response["wind"]["speed"]
        city_cloud= response["clouds"]["all"]
        
        city_data.append({"City": city, "Lat": city_lat, "Lng": city_lng, "Max Temp": city_max_temp,"Humidity": city_humidity, "Wind": city_wind, "Clouds": city_cloud })
        
    except:
        
        print("City Not Found- Skipping")
        
```

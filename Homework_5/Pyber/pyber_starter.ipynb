```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
filepath= "data/city_data.csv"
filepath2="data/ride_data.csv"
city_data= pd.read_csv(filepath)
ride_data= pd.read_csv(filepath2)





# Read the City and Ride Data

# Combine the data into a single dataset

# Display the data table for preview
```

```python
city_ride_data= pd.merge(city_data, ride_data, on=["city"])
city_ride_data.head()

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
urban_cities= city_ride_data[city_ride_data["type"]==("Urban")]
suburban_cities= city_ride_data[city_ride_data["type"]==("Suburban")]
rural_cities= city_ride_data[city_ride_data["type"]==("Rural")]
```

```python
urban_ride_count= urban_cities.groupby("city").count()["ride_id"]
urban_avg_fare=  urban_cities.groupby("city").mean()["fare"]
urban_driver_count= urban_cities.groupby("city").mean()["driver_count"]
                                                        
```

```python
suburban_ride_count= suburban_cities.groupby("city").count()["ride_id"]
suburban_avg_fare=  suburban_cities.groupby("city").mean()["fare"]
suburban_driver_count= suburban_cities.groupby("city").mean()["driver_count"]
```

```python
rural_ride_count= rural_cities.groupby("city").count()["ride_id"]
rural_avg_fare=  rural_cities.groupby("city").mean()["fare"]
rural_driver_count= rural_cities.groupby("city").mean()["driver_count"]
```

```python
plt.title("Pyber Ride Sharing Data 2016")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

plt.scatter(urban_ride_count,urban_avg_fare, s=urban_driver_count, label= "Urban", color= "coral")
plt.scatter(suburban_ride_count,suburban_avg_fare, s=suburban_driver_count, label= "Suburban", color="lightskyblue")
plt.scatter(rural_ride_count,rural_avg_fare, s=rural_driver_count, label= "Rural", color="gold")
plt.legend()
plt.savefig("Pyber_Scatter_2016")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure
```

```python
type_percents=(city_ride_data.groupby("type").sum()["fare"]/city_ride_data["fare"].sum()) * 100

```

```python

explode= (0,0,0.1)
colors = ["gold","lightskyblue","lightcoral"]
city_type= ["Rural", "Suburban", "Urban"]
```

```python
plt.pie(type_percents, explode=explode, labels=city_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")
plt.savefig("Fare_Type_Pie.png")
```

```python
# Show Figure

plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure
```

```python
ride_percents=(city_ride_data.groupby("type").sum()["ride_id"]/city_ride_data["ride_id"].sum()) * 100
```

```python
explode= (0,0,0.1)
colors = ["gold","lightskyblue","lightcoral"]
city_type= ["Rural", "Suburban", "Urban"]
```

```python
plt.pie(ride_percents, explode=explode, labels=city_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")
plt.savefig("Ride_Type_Pie.png")

```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# Build Pie Charts

# Save Figure
```

```python
driver_percents=(city_ride_data.groupby("type").sum()["driver_count"]/city_ride_data["driver_count"].sum()) * 100
```

```python
explode= (0,0,0.15)
colors = ["gold","lightskyblue","lightcoral"]
city_type= ["Rural", "Suburban", "Urban"]
```

```python
# Show Figure
plt.show()
```
